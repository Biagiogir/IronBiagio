```python
num_iter = len(soup.select(".o-chart-results-list__item #title-of-a-story"))
num_iter
```

```python
title = []
artist = []
```

```python
for i in range(num_iter):
    title.append(soup.select(".o-chart-results-list__item #title-of-a-story")[i].get_text(strip=True)) ## getting movies titles
    artist.append(soup.select(".c-label.a-no-trucate")[i].get_text(strip=True))
    
```

```python
len(title)
```

```python
song_rec = pd.DataFrame({"title":title,
                       "artist":artist,
                      })

song_rec.head()
```

```python
def songr():
    user_input = input()
    if user_input in list(song_rec["title"]):
        return np.random.choice(song_rec.title[song_rec["title"]!=user_input])
    else:
        return "Unfortunately, the song is not in the hot list!"
    
```
